# Data Processing & Analysis --

```python
import pandas as pd

df = pd.read_excel('Dataset-PEP.xlsx', sheet_name='Dataset-PEP')
print(df.shape)
df.head()
```

```python
droplist = ['REF_ID', 'AGE CLASS']

data = df.drop(droplist, axis=1)

print(data.shape)
data.head(10)
```

```python
data.info()
```

```python
# Changing categorical data with label:

# JOB :
data['JOB'] = data['JOB'].replace('Employee',1)
data['JOB'] = data['JOB'].replace('Unemployed',2)
data['JOB'] = data['JOB'].replace('Entrepreneur',3)
data['JOB'] = data['JOB'].replace('Housemaid/Wife',4)
data['JOB'] = data['JOB'].replace('Student',5)

# Marital Status :
data['MARITAL_STATUS'] = data['MARITAL_STATUS'].replace('Married',1)
data['MARITAL_STATUS'] = data['MARITAL_STATUS'].replace('Single',2)
data['MARITAL_STATUS'] = data['MARITAL_STATUS'].replace('Divorced',3)

# Education :
data['EDUCATION'] = data['EDUCATION'].replace('High School',1)
data['EDUCATION'] = data['EDUCATION'].replace('University',2)
data['EDUCATION'] = data['EDUCATION'].replace('Below High School',3)

# OWN_CC :
data['OWN_CC'] = data['OWN_CC'].replace('No',1)
data['OWN_CC'] = data['OWN_CC'].replace('Yes',2)

# HOUSE_TYPE :
data['HOUSE_TYPE'] = data['HOUSE_TYPE'].replace('Rent',1)
data['HOUSE_TYPE'] = data['HOUSE_TYPE'].replace('Private House',2)

# CONVERT :
data['CONVERT'] = data['CONVERT'].replace('No',0)
data['CONVERT'] = data['CONVERT'].replace('Yes',1)

data.head()
```

```python
# Checking the data

print('========= ORIGINAL DATA (Size:',data.shape[0], ') =========')
print('CONVERT Case Count:',data.loc[data.CONVERT == 1].shape[0])
print('Non-CONVERT Case Count:',data.loc[data.CONVERT == 0].shape[0])
print('Conversion Rate:', (data.loc[data.CONVERT == 1].shape[0])*100/(data.shape[0]), '%')
```

# Predictive Model's Part --

```python
# Splitting The Data :

from sklearn.model_selection import train_test_split

seed = 42
data_train, data_test = train_test_split(data, test_size=0.3, random_state=seed)

print('\n========= TRAINING DATA (Size:', data_train.shape[0],') =========')
print('CONVERT Case Count:',data_train.loc[data_train.CONVERT == 1].shape[0])
print('Non-CONVERT Case Count:',data_train.loc[data_train.CONVERT == 0].shape[0])
print('Conversion Rate:', (data_train.loc[data_train.CONVERT == 1].shape[0])*100/(data_train.shape[0]), '%')

print('\n========= VALIDATION DATA (Size:', data_test.shape[0],') =========')
print('CONVERT Case Count:',data_test.loc[data_test.CONVERT == 1].shape[0])
print('Non-CONVERT Case Count:',data_test.loc[data_test.CONVERT == 0].shape[0])
print('Conversion Rate:', (data_test.loc[data_test.CONVERT == 1].shape[0])*100/(data_test.shape[0]), '%')
```

```python
x_train = data_train.drop(['CONVERT', 'DELTA_DAYS_PREV_CALL_v2', 'NUMBER_PREVIOUS_ATTEMPT_v2'], axis=1)

x_train.head()
```

```python
x_valid = data_test.drop(['CONVERT', 'DELTA_DAYS_PREV_CALL_v2', 'NUMBER_PREVIOUS_ATTEMPT_v2'], axis=1)

x_valid.head()
```

```python
x_train.info()
```

```python
y_train = data_train[['CONVERT']]

y_train.head()
```

```python
y_valid = data_test[['CONVERT']]

y_valid.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

# membuat model Random Forest - MODEL 1
tree_model = RandomForestClassifier()

# melakukan pelatihan model terhadap data
tree_model.fit(x_train, y_train)

y_pred = tree_model.predict(x_valid)

# Membuat gambar tentang importance
import matplotlib.pyplot as plt

print('====The Importance Variables of Model 1 - Random Forest Classifier====\n')

importance = tree_model.feature_importances_
columns_list = x_train.columns.values

total_weight = importance.sum()
transformed_importance = [(x/total_weight)*100 for x in importance]
transformed_importance_v2 = [round(x,2) for x in transformed_importance]

importance_df = pd.DataFrame()
importance_df['Features'] = columns_list
importance_df['Importance'] = transformed_importance_v2
importance_df.sort_values(by=['Importance'], ascending=True, inplace=True)

# summarize feature importance

print(importance_df.sort_values(by=['Importance'], ascending=False).to_string(index=False))

plt.figure(figsize=(10,10))
plt.barh(importance_df['Features'], importance_df['Importance'])
for i, v in enumerate(importance_df['Importance']):
    plt.text(v + .1, i, str(v), color='red', fontweight='bold')
plt.show()
```

```python
# PREDICTION DATA RESULT:

print('Classes from output:',tree_model.classes_, 'where (0) -> NO dan (1) -> YES')

y_probability = tree_model.predict_proba(x_valid)

cutoff = list(range(1, 101))
predictions = []

for i in cutoff:
    y_preds = []
    for prob in y_probability:
        if (float(prob[1]) > (i/100)):
            y_preds.append(1)
        else:
            y_preds.append(0)
            
    y_pred_df = pd.DataFrame()
    y_pred_df['CONVERT'] = y_preds
    predictions.append(y_pred_df)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print('--- The Confusion Matrix: ---\n\n')
tn, fp, fn, tp = confusion_matrix(y_valid, y_pred).ravel()
print(confusion_matrix(y_valid, y_pred),'\n')

print('True Positive:', tp)
print('True Negative:', tn)
print('False Positive:', fp)
print('False Negative:', fn)

print('\n--- The Test Reports: ---\n',classification_report(y_valid, y_pred))
```

```python
# Print the Cutoff Probabilities and Precision and Accuracy :
from sklearn.metrics import confusion_matrix, classification_report

precisions = []
accuracy = []
recall = []
f_score = []
number_predicted_yes = []

for i in cutoff:
    tn, fp, fn, tp = confusion_matrix(y_valid, predictions[i-1]).ravel()
    accuracy.append(((tp+tn)/(tp+tn+fp+fn))*100)
    recall.append((tp/(tp+fn))*100)
    
    if i == 100:
        precisions.append(0)
        f_score.append(0)
    else:
        precisions.append(((tp/(tp+fp))*100))
        f_score.append((2*((tp/(tp+fn))*100)*(((tp/(tp+fp))*100)))/(((tp/(tp+fn))*100) + (((tp/(tp+fp))*100))))
    
for i in cutoff:
    u = predictions[i-1][predictions[i-1].CONVERT == 1].shape[0]
    number_predicted_yes.append(u)

pred_table = pd.DataFrame()

pred_table["Cutoff_Level (%)"] = cutoff
pred_table["Precision (%)"] = precisions
pred_table["Number_Pred_YES"] = number_predicted_yes
pred_table["Accuracy (%)"] = accuracy
pred_table["Recall (%)"] = recall
pred_table["F1-Score (%)"] = f_score

pred_table.head(10)
```

```python
# Save the Cuttoff sampling Result into excel file :

pred_table.to_excel('Model_PEP_Metrics_Sampling.xlsx')
```

```python
# Create ROC curve and AUC:

from sklearn import metrics

y_pred_proba = tree_model.predict_proba(x_valid)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_valid,  y_pred_proba)
auc = metrics.roc_auc_score(y_valid, y_pred_proba)


plt.plot(fpr,tpr,label="AUC="+str(auc))
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.legend(loc=4)
plt.show()
```

```python
# Scored Data :

y_score = []

for prob in y_probability:
    y_score.append(prob[1]*100)
    
scored_data = x_valid

scored_data['Score'] = y_score

scored_data.sample(25)
```

```python
# Save the Scored Data into excel file :

scored_data.to_excel('Scoring_Data_Results_PEP.xlsx')
```
